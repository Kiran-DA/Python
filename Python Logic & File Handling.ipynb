```python
import pandas as pd  # Import the pandas library for data manipulation
import re  # Import the regular expressions library for input validation

print(f"Maintaining Student Records")  # Print the title of the program

# Initialize an empty dictionary to store student details
students_details = {}
# Initialize an empty list to store student records
students_list = []
# Create an empty DataFrame with specified columns for student records
df = pd.DataFrame(columns=['name', 'rollno', 'm', 'sci', 's'])

# Start an infinite loop to allow continuous user interaction
while True:
    print('-' * 50)  # Print a separator line
    print("1. Add Student Details")  # Option to add student details
    print("2. Display Student records")  # Option to display student records
    print("3. Calculate Average marks")  # Option to calculate average marks
    print("4. Save Students records to a file")  # Option to save records to a file
    print("5. Exit")  # Option to exit the program
    c = int(input("Enter the choice:"))  # Get user choice

    if c == 1:  # If the user chooses to add student details
        name = input("Enter Student Name:")  # Prompt for student name
        # Check if the name contains any digits
        if re.search(r'\d', name):
            print("Name contains numbers")  # Print error message if name is invalid
        else:
            rollno = int(input("Enter Student Roll number:"))  # Prompt for roll number
            # Prompt for marks in different subjects
            math = int(input("Enter Student marks in Maths Subject:"))
            science = int(input("Enter Student marks in Science Subject:"))
            social = int(input("Enter Student marks in Social Subject:"))
            # Update the students_details dictionary with the new student's information
            students_details.update({"name": name, "rollno": rollno, "m": math, "sci": science, "s": social})
            # Append the student details dictionary to the students_list
            students_list.append(students_details)
            # Concatenate the new student details to the DataFrame
            df = pd.concat([df, pd.DataFrame([students_details])], ignore_index=True)

    elif c == 2:  # If the user chooses to display student records
        if students_details:  # Check if there are any student details
            print("Student Details:")  # Print header
            print(df)  # Display the DataFrame containing student records
        else:
            print("No Students Records found")  # Print message if no records exist

    elif c == 3:  # If the user chooses to calculate average marks
        df['average'] = df[['m', 'sci', 's']].mean(axis=1)
        print("Average Score is:")
        print(df)  # Placeholder for average score calculation

    elif c == 4:  # If the user chooses to save student records to a file
        filename = 'output.csv'  # Specify the filename for the CSV file
        # Write the DataFrame to a CSV file
        df.to_csv(filename, index=False)  # Save the DataFrame to the specified CSV file
        print("Student Data has been stored to:", filename)  # Confirm that data has been saved

    else:  # If the user chooses to exit
        break  # Exit the loop and end the program
```
